```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
wine = pd.read_csv('winequality-red.csv', header=None, sep=";", dtype=str)
```

```python
wine.columns = ["fixed acidity", "volatile acidity", "citric acid", "residual sugar", "chlorides", "free sulfur dioxide", "total sulfur dioxide", "density", "pH", "sulphates", "alcohol", "quality"]
```

```python
wine = wine.drop(wine.index[0])
```

```python
wine
```

```python
wine_features = wine.drop("quality", axis = 1)
wine_quality = wine.quality.astype(float)
```

```python
wine_features.head()
```

```python
wine_quality.head()
```

```python
wine_features.dtypes
```

```python
wine_features[['fixed acidity']] = wine_features[['fixed acidity']].astype(float)
wine_features['volatile acidity'] = wine_features['volatile acidity'].astype(float)
wine_features['citric acid'] = wine_features['citric acid'].astype(float)
wine_features['residual sugar'] = wine_features['residual sugar'].astype(float)
wine_features['chlorides'] = wine_features['chlorides'].astype(float)
wine_features['free sulfur dioxide'] = wine_features['free sulfur dioxide'].astype(float)
wine_features['total sulfur dioxide'] = wine_features['total sulfur dioxide'].astype(float)
wine_features['density'] = wine_features['density'].astype(float)
wine_features['pH'] = wine_features['pH'].astype(float)
wine_features['sulphates'] = wine_features['sulphates'].astype(float)
wine_features['alcohol'] = wine_features['alcohol'].astype(float)
```

```python
wine_features.dtypes
```

```python
model = LinearRegression()
```

```python
model.fit(wine_features, wine_quality)
```

```python
model.score(wine_features, wine_quality)
```

```python
model.intercept_
```
